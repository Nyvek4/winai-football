```python
# Importing Libraries
import os
import json
import requests
import time
import pandas as pd
from pandas import json_normalize
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
## PATH

dirname = os.getcwd()

rel_matches_path = 'OneDrive - Association Cesi Viacesi mail\Bureau\Projects\winAi\Datasets\\raw-externe\kaggle\stats-bomb\data\matches'
matches_path = os.path.join(dirname, rel_matches_path)

rel_events_path = 'OneDrive - Association Cesi Viacesi mail/Bureau/Projects/winAi/Datasets/raw-externe/kaggle/stats-bomb/data/events'
events_path = os.path.join(dirname, rel_events_path)
```

```python

def find_json_files(directory):
    json_files = [] 
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith(".json"):
                json_files.append(os.path.join(root, file))
    return json_files

json_files = find_json_files(matches_path)
_matches_df = [pd.read_json(file) for file in json_files]
matches_df = pd.concat(_matches_df, ignore_index=True)

matches_df.reset_index(drop=True, inplace=True)

if(matches_df.empty):
    print("No data found")
else:
    print("Data loaded successfully")
```

```python


def findflat_json_files(directory, limit=None):
    json_files = [os.path.join(directory, file) for file in os.listdir(directory) if file.endswith('.json')]
    if limit:
        json_files = json_files[:limit]
    return json_files

json_files = findflat_json_files(events_path, limit=343)

start_time = time.time()
batch_size = 5  
events_df_list = []

for i in range(0, len(json_files), batch_size):
    batch_files = json_files[i:i+batch_size]
    batch_dfs = []
    for file in batch_files:
        temp_df = pd.read_json(file)
        temp_df['match_id'] = os.path.basename(file).split('.')[0]
        batch_dfs.append(temp_df)
    batch_df = pd.concat(batch_dfs, ignore_index=True)
    events_df_list.append(batch_df)
    print(f'Batch {i//batch_size + 1} loaded.')

events_df = pd.concat(events_df_list, ignore_index=True)
events_df.reset_index(drop=True, inplace=True)

print(f'Total loading time: {time.time() - start_time:.2f} seconds')
print(events_df.head())
```

```python

## DATA PREPROCESSING - EXTRACTION OF STATISTICS FOR EVENTS DATA - BOTH TEAMS

# Extraction des statistiques pour les données d'événements

#--------------------------- Filter on SHOTS ---------------------------#
shots_df = events_df[events_df['type'].apply(lambda x: x['name'] == 'Shot')]


#----------- SHOTS ON TARGET

# Filtrer les tirs pour ne conserver que ceux avec un outcome de "Saved"
shots_on_target_df = shots_df[shots_df['shot'].apply(lambda x: x['outcome']['name'] == 'Saved')]
# Nombre de tirs cadrés sur le but par match
shots_on_target_per_match = shots_on_target_df.groupby('match_id').size()
shots_on_target_per_match = shots_on_target_per_match.rename_axis('match_id').reset_index(name='shots_on_target') 
print(shots_on_target_per_match.head())

#----------- SHOTS OFF TARGET

# Filtrer les tirs pour ne conserver que ceux avec un outcome de "Off T"
shots_off_target_df = shots_df[shots_df['shot'].apply(lambda x: x['outcome']['name'] == 'Off T')]
# Nombre de tirs non cadrés par match
shots_off_target_per_match = shots_off_target_df.groupby('match_id').size()
shots_off_target_per_match = shots_off_target_per_match.rename_axis('match_id').reset_index(name='shots_off_target')
print(shots_off_target_per_match.head())

#----------- GOALS

# Filtrer les tirs pour ne conserver que ceux avec un outcome de "Goal"
goals_df = shots_df[shots_df['shot'].apply(lambda x: x['outcome']['name'] == 'Goal')]
# Nombre de buts par match
goals_per_match = shots_on_target_df.groupby('match_id').size()
goals_per_match = goals_per_match.rename_axis('match_id').reset_index(name='goals')
print(goals_per_match.head())

#----------- KEY PASSES

# Filtrer les tirs pour ne conserver que ceux qui ont une passe clé (key_pass_id)
key_passes_df = shots_df[shots_df['shot'].apply(lambda x: 'key_pass_id' in x and x['key_pass_id'] is not None)]
# Nombre de passes clés par match
key_passes_per_match = key_passes_df.groupby('match_id').size()
key_passes_per_match = key_passes_per_match.rename_axis('match_id').reset_index(name='key_passes')
print(key_passes_per_match.head())

# ----------- xGoals

# Extraire les valeurs de xG pour chaque tir
xG_values = shots_df['shot'].apply(lambda x: x['statsbomb_xg'])
# Ajouter les valeurs xG à la dataframe des tirs
shots_df['xG'] = xG_values
# Somme des valeurs xG par match
xG_per_match = shots_df.groupby('match_id')['xG'].sum()
xG_per_match = xG_per_match.rename_axis('match_id').reset_index(name='xG')
print(xG_per_match.head())

# ----------- average xG per Shot per match

# Nombre de tirs par match
shots_per_match = shots_df.groupby('match_id').size()
# Somme des valeurs xG par match
xG_per_match = shots_df.groupby('match_id')['xG'].sum()
# xG moyen par tir par match
average_xG_per_shot_per_match = xG_per_match / shots_per_match
average_xG_per_shot_per_match = average_xG_per_shot_per_match.rename_axis('match_id').reset_index(name='average_xG_per_shot')
print(average_xG_per_shot_per_match.head())

#--------------------------- Filter on DRIBBLES ---------------------------#
dribbles_df = events_df[events_df['type'].apply(lambda x: x['name'] == 'Dribble')]

# ----------- succesfull dribbles
# Filtrer les dribbles réussis
successful_dribbles_df = dribbles_df[dribbles_df['dribble'].apply(lambda x: x['outcome']['name'] == 'Complete')]
# Nombre de dribbles réussis par match
dribbles_per_match = successful_dribbles_df.groupby('match_id').size()
dribbles_per_match = dribbles_per_match.rename_axis('match_id').reset_index(name='successful_dribbles')
print(dribbles_per_match.head())

# ----------- unsuccessful dribbles

# Filtrer les dribbles non réussis
unsuccessful_dribbles_df = dribbles_df[dribbles_df['dribble'].apply(lambda x: x['outcome']['name'] == 'Incomplete')]
# Nombre de dribbles non réussis par match
unsuccessful_dribbles_per_match = unsuccessful_dribbles_df.groupby('match_id').size()
unsuccessful_dribbles_per_match = unsuccessful_dribbles_per_match.rename_axis('match_id').reset_index(name='unsuccessful_dribbles')
print(unsuccessful_dribbles_per_match.head())

#--------------------------- Filter on FOULS ---------------------------#

# Filtrer les événements de type "Foul Won"
fouls_won_df = events_df[events_df['type'].apply(lambda x: x['name'] == 'Foul Won')]
fouls_won_df = fouls_won_df[fouls_won_df['foul_won'].apply(lambda x: isinstance(x, dict) and 'defensive' in x and x['defensive'] is not None)]
# Nombre de fautes gagnées par match
fouls_won_per_match = fouls_won_df.groupby('match_id').size()
fouls_won_per_match = fouls_won_per_match.rename_axis('match_id').reset_index(name='fouls_won')
print(fouls_won_per_match.head())
```
